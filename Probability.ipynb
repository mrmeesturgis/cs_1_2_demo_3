```python
def dicepair():
    r1 = random.randint(1,6)
    r2 = random.randint(1,6)
    return r1 + r2
```

```python
def xrolls(rolls: int):
    r1 = list(range(2,13))
    d1 = []
    d2 = []
    results = {}
    for n in r1:
        results[n] = 0
    while rolls > 0:
        rolls -= 1
        results[dicepair()] += 1
    for k, v in results.items():
        d1.append(k)
        d2.append(v)
    return d1, d2
```

```python
d1, d2 = xrolls(10)
print(d1, d2)
```

```python
fig, ax = plt.subplots(figsize=(10, 5.4))
ax.scatter(d1, d2, s=50, facecolor='C0', edgecolor='k');
ax.stem(d1, d2, linefmt='r')
ax.plot(d1, d2)
```

```python
def xrolls_human(rolls: int):
    r1 = list(range(2,13))
    d1 = []
    d2 = []
    results = {}
    for n in r1:
        results[n] = 0
    while rolls > 0:
        rolls -= 1
        results[int(input('what did you roll? '))] += 1
    for k, v in results.items():
        d1.append(k)
        d2.append(v)
    return d1, d2
```

```python
hd1, hd2 = xrolls_human(10)
```

```python
print(hd1, hd2)
```

```python
fig, ax = plt.subplots(figsize=(10, 5.4))
ax.scatter(hd1, hd2, s=50, facecolor='C0', edgecolor='k');
ax.stem(hd1, hd2, linefmt='r')
ax.plot(hd1, hd2)
```
